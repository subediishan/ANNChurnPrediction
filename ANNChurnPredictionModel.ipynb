```python
import pandas as pd
import tensorflow.keras as keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import numpy as np

```

```python
import pickle as pk
```

```python
data = pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis = 1 )
```

```python
data.head(3)
```

```python
#encode categorical data converts gender male female to 0 or 1 
label_encoder_gender = LabelEncoder()
x = label_encoder_gender.fit_transform(data['Gender']) 
data['Gender'] = x
data.head(3)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
```

```python
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
frame  = pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out())
```

```python
data = pd.concat([data.drop('Geography',axis= 1 ),frame],axis = 1 )
data
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pk.dump(label_encoder_gender,file)
```

```python
with open ('onehot_encoder_geo.pkl', 'wb') as file:
  pk.dump(onehot_encoder_geo,file)
```

```python
x = data.drop('Exited', axis = 1)
y = data.drop(x.columns,axis = 1)
```

```python
x_train, x_test , y_train,y_test = train_test_split(x , y , random_state = 42, test_size = 0.2)
x_train
```

```python
scalar =  StandardScaler()
x_train = scalar.fit_transform(x_train)
x_train
```

```python
with open ('scalar.pkl','wb') as file:
    pk.dump(scalar,file)
    

```

```python
import tensorflow as tf
from  tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

```

```python
model = Sequential([
  Dense(64, activation='relu' , input_shape = ((x_train.shape[1],) )),
  Dense(32,activation= 'relu' ),
  Dense(1,activation='sigmoid')
])
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate = 0.1)
loss = tensorflow.keras.losses.BinaryCrossentropy()

model.compile(optimizer=opt,loss = loss , metrics=['accuracy'])
```

```python
import datetime

log_dir = "logs/fit"+datetime.datetime.now().strftime("%y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir = log_dir, histogram_freq= 1 )
early_stopping_callbacks = EarlyStopping(monitor= 'val_loss' , patience= 5 , restore_best_weights= True)
```

```python
history = model.fit(
  x_train,y_train, validation_data= (x_test,y_test),epochs = 100, callbacks = [tensorflow_callback,early_stopping_callbacks]
)
```

```python
model.save('model.keras')
```
