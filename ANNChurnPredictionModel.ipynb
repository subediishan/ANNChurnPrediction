```python
import pandas as pd
import tensorflow.keras as keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import numpy as np

```

```python
import pickle as pk
```

```python
data = pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis = 1 )
```

```python
data.head(3)
```

```python
#encode categorical data converts gender male female to 0 or 1 
label_encoder_gender = LabelEncoder()
x = label_encoder_gender.fit_transform(data['Gender']) 
data['Gender'] = x
data.head(3)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
```

```python
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
frame  = pd.DataFrame(one_hot_encoder_geo.toarray(),columns=one_hot_encoder.get_feature_names_out())
```

```python
data = pd.concat([data.drop('Geography',axis= 1 ),frame],axis = 1 )
data
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pk.dump(label_encoder_gender,file)
```

```python
with open ('onehot_encoder_geo.pkl', 'wb') as file:
  pk.dump(onehot_encoder_geo,file)
```
